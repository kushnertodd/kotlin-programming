*Data* is measured quantities or qualities, observations, facts, or statistics of units of meaning or symbols.
Data can be seen as the smallest units of factual information
that can be used as a basis for calculation, reasoning, or discussion.
Data is collected using techniques such as measurement, observation, query, or analysis,
Once data is collected, it only becomes information suitable for making decisions once it has been analyzed and interpreted.

## Levels of Data
Data values have four levels of measurement.
There are two general categories of data.

- Qualitative or discrete values that can be counted and has a limited number of values.
- Quantitative or continuous values that can be measured and can take on an unlimited number of different values.

These are the specific levels of data and actions that can be performed on them.

Level|Type|Description |Actions|Examples
:- |:-|:-|:-|:-
Nominal |Qualitative|Values are classified into labels or categories|Count|Colors, countries, blood type
Ordinal |Qualitative|Values rank higher or lower than others |Sort|Income range (low, medium high), grades, skill level
Interval |Quantitative|Values have measures of difference between them, but do not have a 0 origin|Take difference|Temperature, time
Ratio |Quantitative|Values have proportions of each other on a scale that has a 0 origin|Compute ratio|Weight, length, duration
