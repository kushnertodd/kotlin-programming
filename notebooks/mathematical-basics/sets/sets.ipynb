## Sets
*Sets* are a group of unique elements.
There can be any number and any type, as long as each is distinct from any other.
Each element is a *member* of the set.

A set contains some number of distinct elements.
The set `S` contains the elements `a`, `b`, `c`, `d`.

<img src="sets-S.jpg" width="500"/>

A set is expressed using curly braces `{` and `}`.

`S = {a, b, c, d}`

<a id="set-membership"></a>
### Set membership

We say that a member is an *element* of a set using the $\in$ operator.
         
<img src="sets-a-elem-S.jpg" width="500"/>

A set can be a *subset* of another set if all its elements are members of the other set.

<img src="sets-P-sub-S.jpg" width="500"/>

<a id="set-Union"></a>
### Set Union
When sets are combined, the results only include the unique elements from both sets.
The *union* of two sets are all the members from each set with duplicates removed. 
Here,

`S = {a, b, c, d}`  
`T = {c, d, e, f}`  
`S`$\cup$`T = {a, b, c, d, e, f}`  

<img src="sets-S-union-T.jpg" width="500"/>

<a id="set-Intersection"></a>
### Set Intersection
The *intersection* of two sets are all the members that are common to both sets. 
Here,

`S = {a, b, c, d}`  
`T = {c, d, e, f}`  
`S`$\cap$`T = {c, d}`  

<img src="sets-S-inter-T.jpg" width="500"/>

<a id="set-addition"></a>
### Set addition
Elements can be added to sets.
Here, the element `b` is added to the set `P = {a}`, giving the set `P + b = {a, b}`.

<img src="sets-P-plus-b.jpg" width="500"/>

<a id="set-removing-elements"></a>
### Removing elements from sets
Elements can be removed from sets.
Here, the element `b` is removed from the set `R = {a, b}`, giving the set `R - b = {a}`.

<img src="sets-R-less-b.jpg" width="500"/>

<a id="set-removing-sets"></a>
### Removing sets from sets
Sets can be removed from sets.
Here, the set `P = {a}` is removed from the set `S = {a, b, c, d}`, giving the set `S - P = {b, c, d}`.

<img src="sets-S-less-P.jpg" width="500"/>
